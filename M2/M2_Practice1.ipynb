**El tiempo máximo sin promoción de las personas que llevan más de 15 años de servicio**

```python
maxtime_wo_promotion = 0
time_wo_promotion = dfemployee_attrition[['Employee_ID','Time_of_service','Time_since_promotion']]
time_wo_promotion = time_wo_promotion[time_wo_promotion['Time_of_service']>15]
#time_wo_promotion = time_wo_promotion.sort_values('Time_since_promotion', ascending=False)
maxtime_wo_promotion = time_wo_promotion['Time_since_promotion'].max()
maxtime_wo_promotion
```

**La edad promedio en la que ingresaron las personas a la empresa, genera una nueva columna derivada mediante la formula Age - Time_of_service**

```python
adm_average = 0
average_admission = dfemployee_attrition[['Age','Time_of_service']].copy()
average_admission['adm_date'] = dfemployee_attrition['Age'] - dfemployee_attrition['Time_of_service']
average_admission

#dfemployee_attrition
adm_average = average_admission['adm_date'].mean().round(2)
adm_average
```

**Un dataframe con las columnas Employee_ID, Relationship, Decision_skill_process que sólo contenga los registros de mujeres, solteras y con compensación tipo 2 o 3**

```python
women_analysis = dfemployee_attrition[['Employee_ID','Gender', 'Relationship_Status', 'Decision_skill_possess','Compensation_and_Benefits']]
women_analysis = women_analysis[women_analysis['Gender'] == 'F']
women_analysis = women_analysis[women_analysis['Relationship_Status'] == 'Single']
women_analysis = women_analysis[(women_analysis['Compensation_and_Benefits'] == 'type2') | (women_analysis['Compensation_and_Benefits'] == 'type3')]
print('Tamaño : ', women_analysis.shape)
women_analysis
```

**Una columna derivada con el valor "on time" si cuando tuvo su última promoción tenía menos de 30 años. En caso contrario colocar el valor de "out of time"**

```python
last_promotion = dfemployee_attrition[['Age','Time_since_promotion']].copy()

# Option 1
last_promotion['promotion_age_ok'] = np.where(
      (last_promotion['Age'] - last_promotion['Time_since_promotion'] >= 30), 'out of time',
          np.where((last_promotion['Age'] - last_promotion['Time_since_promotion']<30), 'on time', 'NA'
        )
    ) 
last_promotion

# Option 2
#last_promotion['promotion_age_ok'] = 'NA'
#last_promotion.loc[(last_promotion['Age'] - last_promotion['Time_since_promotion'] >= 30), 'promotion_age_ok'] = 'out of time'
#last_promotion.loc[(last_promotion['Age'] - last_promotion['Time_since_promotion'] < 30), 'promotion_age_ok'] = 'out time'
#last_promotion

#Result check

#valprom_ok = last_promotion[['Age','Time_since_promotion','promotion_age_ok']]
#valprom_ok = valprom_ok['promotion_age_ok'].value_counts()
#valprom_ok


```
