# Crear 4 DataFrames en pandas, cada uno para cargar el contenido de los diferentes archivos.

```python
#get data from csv files
employees_one = pd.read_csv('Test.csv')
employees_one.info()

employees_two = pd.read_csv('Train.csv')
employees_two.info()


additional_employees_one = pd.read_csv('Test2.csv')
additional_employees_one.info()

additional_employees_two = pd.read_csv('Train2.csv')
additional_employees_two.info()
```

# Conjuntar de manera correcta la información, de tal forma que queden todos los registros juntos en un solo DataFrame que contenga todas las columnas necesarias

```python
employees =  pd.concat([employees_one,employees_two], ignore_index=True)
employees.shape

additional_employees = pd.concat([additional_employees_one,additional_employees_two], ignore_index=True)
additional_employees.shape
```

```python
employees_info = employees.merge(additional_employees, left_on='Employee_ID', right_on='Employee_ID', how='left', suffixes=('_emp','_add'))
#employees_info.info()
#employees_info.shape
#employees_info

#rename of column names with spaces
#employees_info.rename(columns={' Computer_OS':'Computer_OS','   Computer_OS_version':'Computer_OS_version',' Computer_tickets':'Computer_tickets',' Mobile_OS':'Mobile_OS',' Mobile_OS_version':'Mobile_OS_version',' Mobile_tickets':'Mobile_tickets'})
employees_info.columns = employees_info.columns.str.lstrip()
employees_info.info()
```

# Para todos los registros que no tienen un valor para un sistema operativo, ya sea de su computadora o de su dispositivo móvil, asignar el valor “no egistrado"

```python
#employees_info.loc[employees_info['Computer_OS']=='Windows', ['Computer_OS']] = 'no registrado'
employees_info.loc[employees_info['Computer_OS'].isnull(),'Computer_OS']  = 'no registrado'
employees_info.loc[employees_info['Computer_OS'] == 'no registrado']

employees_info.loc[employees_info['Mobile_OS'].isnull(),'Mobile_OS']  = 'no registrado'
employees_info.loc[employees_info['Mobile_OS'] == 'no registrado']
```

# El porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora

```python
percentage_os = employees_info[['Computer_OS','Employee_ID']]
dfpercentage_os = percentage_os.groupby(['Computer_OS'], as_index=False)['Employee_ID'].count()
dfpercentage_os['percentage'] = (dfpercentage_os['Employee_ID'] / dfpercentage_os['Employee_ID'].sum())*100
dfpercentage_os
```

# El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo móvil

```python
percentage_os_mobile = employees_info[['Mobile_OS','Employee_ID']]
dfpercentage_os_mobile = percentage_os_mobile.groupby(['Mobile_OS'], as_index=False)['Employee_ID'].count()
dfpercentage_os_mobile['Promedio'] = (dfpercentage_os_mobile['Employee_ID'] / dfpercentage_os_mobile['Employee_ID'].sum())*100
dfpercentage_os_mobile
```

# La edad promedio de la gente que utiliza cada sistema operativo de escritorio

```python
#average_age = employees_info[['Computer_OS','Age','Employee_ID']]
#dfaverage_age = average_age.groupby(['Computer_OS','Age'], as_index=False)['Employee_ID'].count()
#average_age_by_so = dfaverage_age.groupby('Computer_OS', as_index=False)['Age'].mean().round()
#average_age_by_so

avg_age = employees_info[['Computer_OS','Age']]
average_age_by_os = avg_age.groupby(['Computer_OS']).agg('mean').round()
average_age_by_os
```

# El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio

```python
average_el = employees_info[['Computer_OS','Education_Level']]
education_level_by_os = average_el.groupby(['Computer_OS']).agg('mean').round()
education_level_by_os
```

# El número total de tickets generados por la gente con cada sistema operativo de escritorio

```python
tickets_os = employees_info[['Computer_OS','Computer_tickets']]
total_tickets_os = tickets_os.groupby(['Computer_OS'], as_index=False)['Computer_tickets'].sum()

#total_tickets_os['percentage'] = (dfpercentage_os['Employee_ID'] / dfpercentage_os['Employee_ID'].sum())*100
total_tickets_os
```

```python
total_tickets_os = tickets_os.groupby(['Computer_OS']).agg('sum')
total_tickets_os
```
